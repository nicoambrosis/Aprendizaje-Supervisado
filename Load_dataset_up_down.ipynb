Vamos a generar las etiquetas de los datos. 

Como queremos que los datos sean linealmente separables vamos a establecer como condicion que aquellos datos que tengan un valor de Y superior a 0 queden etiquetados como '1' y los que tengan un valor de Y inferior o igual a 0 quedaran etiquetados como '0'

```python
Y = (X[:,1] > 0).astype(np.int8)
Y
```

### `load_dataset_up_down`
Definimos la funcion y luego la aplicamos.

```python
def load_dataset_up_down(size, seed=39):
    np.random.seed(seed)
    x = np.random.poisson(5, size) * (np.random.randint(0,2, size) * 2 - 1)
    y = (np.random.poisson(5, size) + 1) * (np.random.randint(0,2, size) * 2 - 1)
    X = np.array(list(zip(x,y)))
    Y = (X[:,1] > 0).astype(np.int8)
    return X, Y
```

```python
X, Y = load_dataset_up_down(100, seed = 39)
print(' Datos '. center(100, '='))
display(X[:5])
print(' Etiquetas '.center(100,'='))
display(Y)
```

### Visualizamos el resultado

```python
import matplotlib.pyplot as plt
```

```python
indices = np.argsort(Y)
X = X[indices]
Y = Y[indices]

plt.scatter(x = X[:, 0], 
            y = X[:, 1],
            c=Y,
            s=50,
            edgecolors = 'k',
            cmap=plt.cm.plasma)

plt.xlabel('x')
plt.ylabel('y')

plt.grid(alpha = 0.50)

plt.show()
```
